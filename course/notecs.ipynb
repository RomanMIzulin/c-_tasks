```python
from os import listdir
import sys
# bin(int.from_bytes(t.encode(), 'little'))
def readOne(file):
    lis = list()
    with open(file,'rb') as f:
        lis=[]
        while(True):
            t = f.read(1)
            
            if t!=b'':
                lis.append(t)
            else:
                break
    return lis

def readFiles(direc):
    d = dict()
    fileNames =  [f for f in listdir(direc)]
    for i in fileNames:
        d.update({i:readOne(direc+'/'+i)})
    return d

def analyseOne(lis):
    s = sorted(set(lis))
    d = dict()
    for i in s:
        count = 0
        for j in lis:
            if j<i:
                count = count+1
        d.update({i:count})
    return d

def mean(lis):
    sum = 0
    for i in lis:
        sum +=int.from_bytes(i,byteorder='little')
    return sum/len(lis)


def desp(lis):
    sum = 0
    for i in lis:
        sum +=(int.from_bytes(i,byteorder='little')**2)
    return sum/len(lis)-mean(lis)**2

def showAll(dic):
    ordered  = sorted(dic.items(), key=lambda kv:kv[0])
    for i in ordered:
        print(ordered[0],' ',format(mean(ordered[1]),'.2e'), ' ', format(desp(ordered[1]),'.2e'))
#     for key,value in dic.items():
#         print(key,' ',format(mean(value),'.2e'), ' ', format(desp(value),'.2e'))
    
    
print("Native byteorder: ", sys.byteorder)

# print(readOne('./pics/'+l[0]))
# readOne('pics'+'/'+'eee')
data = readFiles('testfiles')
# data['first.png'][0]==data['first.png'][0]
# analyseOne(data['eee'])
# mean(data['first.png']),mean(data['second.png']),mean(data['eee']),desp(data['first.png']),desp(data['eee'])
# showAll(data)
data
```
